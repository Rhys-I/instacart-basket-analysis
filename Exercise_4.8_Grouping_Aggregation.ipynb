## 1Ô∏è‚É£ Import Libraries & Data

```python
import pandas as pd
import numpy as np
import os

# Set path
path = r'C:\Users\rhysm\OneDrive\Desktop\Career Foundry\Data Immersion\Module 4\04-2025 Instacart Basket Analysis'

# Import dataset
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_merged_labeled.pkl'))

# Confirm import
ords_prods_merge.head()
```

## 3Ô∏è‚É£ Grouping and Aggregating (Entire Dataset)

```python
# Mean of 'order_number' grouped by 'department_id'
dept_order_mean = ords_prods_merge.groupby('department_id')['order_number'].mean()
print(dept_order_mean)
```

## üí¨ Comment

## 4Ô∏è‚É£ Loyalty Flag

```python
# Create max_order column
ords_prods_merge['max_order'] = ords_prods_merge.groupby('user_id')['order_number'].transform(np.max)

# Create loyalty_flag
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'

# Check distribution
ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

## 5Ô∏è‚É£ Spending Habits by Loyalty

```python
# Descriptive stats of prices by loyalty_flag
ords_prods_merge.groupby('loyalty_flag')['prices'].describe()
```

## 6Ô∏è‚É£ Spending Flag

```python
# Create avg_price column
ords_prods_merge['avg_price'] = ords_prods_merge.groupby('user_id')['prices'].transform(np.mean)

# Create spending_flag
ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spending_flag'] = 'Low spender'
ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spending_flag'] = 'High spender'

# Check distribution
ords_prods_merge['spending_flag'].value_counts(dropna=False)
```

## 7Ô∏è‚É£ Order Frequency Flag

```python
# Create median_days_since column
ords_prods_merge['median_days_since'] = ords_prods_merge.groupby('user_id')['days_since_prior_order'].transform(np.median)

# Create order_frequency_flag
ords_prods_merge.loc[ords_prods_merge['median_days_since'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_days_since'] > 10) & (ords_prods_merge['median_days_since'] <= 20), 'order_frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_days_since'] <= 10, 'order_frequency_flag'] = 'Frequent customer'

# Check distribution
ords_prods_merge['order_frequency_flag'].value_counts(dropna=False)
```
